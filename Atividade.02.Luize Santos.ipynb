```python
# Passo 1: As bibliotecas necessarias P

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

1° Ponto: 1. Quantidade de veículos recuperados por município.

Criando um novo script em Python ou adapte o código já utilizado na análise de roubo de veículos.

```python
# Passo 2: Carregamentos dos dados 

df = df = pd.read_csv('03.BaseDPEvolucaoMensalCisp.csv', encoding= 'iso-8859-1', sep=';')
```

```python
# Passo 3:Colunas necessárias

recuperacao = df[['munic', 'recuperacao_veiculos']]
df.columns
```

```python
# Passo 4: Agrupar por município

recuperacao_municipio = recuperacao.groupby('munic')['recuperacao_veiculos'].sum()
recuperacao.head

# groupby('munic') → agrupa por município

# sum() → soma as recuperações
```

```python
#Passo 5: Ordenar do maior para o menor

recuperacao_municipio = recuperacao_municipio.sort_values(ascending=False)
#Isso cria um ranking.
```

```python
# Passo 6: Mostrar os 10 primeiros

print(recuperacao_municipio.head(10))

#head(10) mostra os 10 maiores valores.
```

Perguntas Respondidas: 

1- Existe grande diferença entre os primeiros municípios do ranking?
Resposta: Sim, normalmente os primeiros municípios apresentam valores muito superiores aos demais, pois grandes
 cidades possuem maior volume de ocorrências e também mais operações policiais que resultam na recuperação de veículos.

2- Compare com o ranking de roubo de veículos. O município que mais tem roubos também lidera em recuperações?

Resposta: Observando que o município do Rio de Janeiro possui um número muito superior de roubos em comparação aos demais.
Entendo que geralmente o município com maior número de roubos também apresenta maior número de recuperações. 
Isso ocorre porque cidades com maior população e maior volume de veículos tendem a concentrar mais ocorrências policiais. 
lém disso, nesses municípios há maior presença de operações policiais voltadas para a recuperação de veículos roubados.


2º Ponto: 2-Análise por ano

Investigação de qual ano teve mais recuperações de veículos.


```python
# Ponto 2 ->>

# 1º Passo: Utilizei novamente o dataframe original.

ano_recuperacao = df[['ano', 'recuperacao_veiculos']]
```

```python
# Ponto 2 ->>

# 2º Passo: Selecionei as colunas: ano e recuperacao_veiculos.


recuperacao_por_ano = ano_recuperacao.groupby('ano')['recuperacao_veiculos'].sum()

```

```python
# Ponto 2 ->>

# 3º Passo:Agrupei os dados por ano.

recuperacao_por_ano = recuperacao_por_ano.sort_values(ascending=False)
```

```python
# Ponto 2 ->>

# 4º Passo: Somei o total de veículos recuperados por ano.
# soma foi feita no passo de agrupamento por ano

print(recuperacao_por_ano)
```

Perguntas Respondidas: 

1- Qual foi o ano com maior número de recuperações de veículos?
Respostas: O ano com maior numero de recuperação foi 2017, com 37.630 veículos recuperados e apresentando o maior numero total.

2-Existe alguma tendência de crescimento ou queda ao longo dos anos?
Respostas: Observando os dados, percebe-se que houve um aumento nas recuperações até aproximadamente 2017, que representa o pico da série. Após esse período, nota-se uma tendência de queda gradual nas recuperações, especialmente a partir de 2020.

3- Esse comportamento acompanha o crescimento ou queda dos roubos de veículos?

R: Apresenta o crescimento dos roubos.Em geral, o comportamento das recuperações tende a acompanhar o número de roubos de veículos. Quando há maior quantidade de roubos, também há maior número de veículos recuperados, pois aumenta o volume de ocorrências investigadas pelas autoridades policiais.




*IMPORTANTE*

O ano de 2017 destaca-se como um outlier positivo, apresentando o maior número de recuperações da série histórica analisada.

3º Ponto: 3- Visualização (Gráfico)

Criei um gráfico de barras mostrando o total de veículos recuperados por ano.



```python
recuperacao_por_ano.plot(kind='bar')

plt.title('Veículos Recuperados por Ano')
plt.xlabel('Ano')
plt.ylabel('Quantidade de Recuperações')

plt.show()
```
