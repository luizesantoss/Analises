1º passo: Carregar os dados

```python
import pandas as pd
```

```python
import numpy as np
```

```python
dados_descontos = pd.read_csv('01.amazon_sales_dataset.csv')
```

```python
dados_descontos.head()
```

```python
dados_descontos.columns
```

2° passo: Isolar a coluna de desconto

```python
descontos = dados_descontos['discount']
```

```python
descontos = descontos.dropna()
```

3º passo : Medidas Centrais

```python
# Calcular Media 

media = descontos.mean()
print('Média: ', media)
```

```python
# Calcular mediana(Q2)

mediana =  descontos.median()
print('Mediana: ', mediana)
```

4º passo: Medidas de Posição

```python
# Calcular Máximo

maximo = descontos.max()
print('Máximo: ', maximo) 
```

```python
# Calcular Mínimo 

minimo = descontos.min()
print('Mínimo: ', minimo)
```

```python
# Calcular Q1 (percentil 25)

q1 = np.percentile (descontos, 25)
print('Q1: ', q1)
```

```python
# Calcular Q3 (percentil 75)

q3 = np.percentile (descontos, 75)
print('Q3: ', q3)
```

```python
# Calcular AMPLITUDE 

amplitude = maximo -  minimo
print('Amplitude: ', amplitude)
```

5º passo :  Assimetria

```python
assimetria = descontos.skew()
print('Assimetria: ', assimetria)
```


6º passo: Desafio de Interpretação

Como interpretar:

Se for ≈ 0 → distribuição simétrica

Se for > 0 → cauda à direita (muitos descontos baixos, poucos muito altos)

Se for < 0 → cauda à esquerda

```python
# Maior e menor desconto

print(f'Maior desconto: $ {maximo:,.2f}')
print (f'Menor desconto: $ {minimo:,.2f}')
```

```python
delta = media - mediana
print("Diferença (Média - Mediana):", delta)
```

```python
razao = media / mediana
print("Razão (Média/Mediana):", razao)
```
